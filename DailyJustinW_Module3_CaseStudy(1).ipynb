```python
"""
Programmer: Justin Wayne Daily
Name of Program: DailyJustinW_Module3_CaseStudy
Description: This program is used to store the attributes for an object of the Automobile class.  The Automobile class is a 
child class of the Vehicle class.  The user is prompted to enter the values for each attribute of the object.  The object's 
attributes are output to the user at the end of the program.

List of variables:
Vehicle - the parent class for the Automobile class.  Represents all potential types of vehicles.
    vehicleType - attribute of the Vehicle class.  Represents the type of vehicle for a Vehicle object.

Automobile - the child class for the Vehicle class.  Represents all potential types of automobiles.
    year - attribute of the Automobile class.  Represents the release year for an Automobile object.
    make - attribute of the Automobile class.  Represents the make for an Automobile object.
    model - attribute of the Automobile class.  Represents the model for an Automobile object.
    doors - attribute of the Automobile class.  Represents the number of doors for an Automobile object.
    roof - attribute of the Automobile class.  Represents the roof type for an Automobile object.

car - an object of the Automobile class, whose first attribute for vehicleType is stored under the Vehicle class.
"""

#Create Vehicle class for all vehicle types
#Vehicle class is the parent class of Automobile class
class Vehicle:
    def __init__(self, vehicleType): #Define the main attributes for a Vehicle object
        """Constructor creates a vehicle with a vehicle type"""
        self.vehicleType = vehicleType

#Create Automobile class for all automobile types
#Make Automobile class a child class of Vehicle
class Automobile(Vehicle): #Define the main attributes for an Automobile object
    def __init__(self, vehicleType, year, make, model, doors, roof):
        """Constructor creates an automobile with a vehicle type, year, make, model, number of doors, and roof type"""
        super().__init__(vehicleType) #Set vehicleType attribute to vehicleType attribute of parent Vehicle class
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

vehicleType = input("Enter the type of vehicle: ")
car = Vehicle(vehicleType)
year = input("Enter the year of the automobile: ")
car.year = year
make = input("Enter the make of the automobile: ")
car.make = make
model = input("Enter the model of the automobile: ")
car.model = model
while True:
    doors = input("Enter the number of doors for the automobile: ")
    if doors == "2" or doors == "4":
        car.doors = doors
        break
while True:
    roof = input("Enter the roof type of the automobile: ")
    if roof == "solid" or roof == "sun roof":
        car.roof = roof
        break
print("")

print(f"Vehicle type: {car.vehicleType}\n" \
      f"Year: {car.year}\n" \
      f"Make: {car.make}\n" \
      f"Model: {car.model}\n" \
      f"Number of doors: {car.doors}\n" \
      f"Type of roof: {car.roof}\n")
```
